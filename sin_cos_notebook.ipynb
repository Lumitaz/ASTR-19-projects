# Sin and Cos Tabulation Notebook

This notebook defines functions for computing **sin(x)** and **cos(x)**, tabulates their values over the interval $[0, 2]$ using NumPy and Astropy, and prints the first 10 entries in a formatted table.

---

## Cell 1 — `compute_sin(x)`

The function below takes a numeric value (or NumPy array) `x` and returns the **sine of x`  `numpy.sin()`. The sine function is a trigonometric function defined as the ratio of the opposite side to the hypotenuse in a right triangle, and oscillates between $-1$ and $1$ with a period of $2\pi$.

```python
import numpy as np

def compute_sin(x):
    """Return the sine of x (in radians). Accepts a scalar or NumPy array."""
    return np.sin(x)
```

## Cell 2 — `compute_cos(x)`

The function below takes a numeric value (or NumPy array) `x` and returns the **cosine** of `x` using `numpy.cos()`.

```python
def compute_cos(x):
    """Return the cosine of x (in radians). Accepts a scalar or NumPy array."""
    return np.cos(x)
```

## Cell 3 — Tabulating sin(x) and cos(x)

Here we use `numpy.linspace()` to generate 1000 spaces between 0.0 and 2 then call our `compute_sin()` and `compute_cos()` functions to build a full table. The result is stored in an `astropy.table.Table` 

```python
from astropy.table import Table

x_values = np.linspace(0.0, 2.0, 1000)
sin_values = compute_sin(x_values)
cos_values = compute_cos(x_values)

trig_table = Table([x_values, sin_values, cos_values], names=("x", "sin(x)", "cos(x)"))

for col in trig_table.colnames:
    trig_table[col].format = ".10f"

print("Table created successfully with", len(trig_table), "entries.")
trig_table
```

## Cell 4 — Printing the First 10 Values

Using a **for loop**, we iterate over the first 10 rows of our table and print $x$, $\sin(x)$, and $\cos(x)$ in vcols, with good formatting

```python
print(f"{'x':>14}  {'sin(x)':>14}  {'cos(x)':>14}")
print("-" * 46)

for i in range(10):
    x   = trig_table["x"][i]
    sinx = trig_table["sin(x)"][i]
    cosx = trig_table["cos(x)"][i]
    print(f"{x:>14.8f}  {sinx:>14.8f}  {cosx:>14.8f}")
```
