```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 100)

def f(x, a, b):
    return a * np.exp(-b * x)

a = 2.0
b = 0.5
```

```python
y = f(x, a, b)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y = a * exp(-b * x)")

plt.savefig("exponential_decay.pdf")
plt.show()
```
